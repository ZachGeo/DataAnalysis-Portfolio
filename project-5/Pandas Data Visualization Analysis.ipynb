# Pandas Data Visualization Analysis

Use **df** to replicate the following plots. 

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('df')
```

```python
df.info()
```

```python
df.head()
```

**Create this scatter plot of b vs a.**

```python
df.plot.scatter(x='a', y='b', color='brown', figsize=(10,3), s=30)
```

**Create a histogram of the 'a' column.**

```python
plt.style.use('ggplot')
df['a'].hist(bins=30, alpha=0.5)
```

**Create a boxplot comparing the a and b columns.**

```python
df.boxplot(['a', 'b'])
```

```python
df[['a', 'b']].boxplot()
```

**Create a kde plot of the 'd' column.**

```python
df['d'].plot.kde()
```

**Figure out how to increase the linewidth and make the linestyle dashed. (Note: You would usually not dash a kde plot line)**

```python
df['d'].plot.kde(lw=5, ls='dashed')
```

**Create an area plot of all the columns for just the rows up to 30. (hint: use .iloc).**
